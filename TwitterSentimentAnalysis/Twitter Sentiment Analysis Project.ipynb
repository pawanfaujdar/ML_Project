```python
import numpy as np
import pandas as pd
```

```python
data_train=pd.read_csv("0000000000002747_training_twitter_x_y_train.csv")
data_test=pd.read_csv("0000000000002747_test_twitter_x_test.csv")
```

```python
data_train
```

```python
categories=data_train["airline_sentiment"]
categories
```

```python
tweets_train=data_train["text"]
tweets_train
```

```python
tweets_test=data_test["text"]
tweets_test
```

```python
from nltk.tokenize import word_tokenize
words_train=[]
for tweet in tweets_train:
    words_train.append(word_tokenize(tweet))
words_train[0]
```

```python
words_test=[]
for tweet in tweets_test:
    words_test.append(word_tokenize(tweet))
words_test[0]
```

```python
from nltk.corpus import stopwords
```

```python
stop=set(stopwords.words("english"))
import string
punctuation=list(string.punctuation)
stop.update(punctuation)
stop
```

```python
documents_train=np.c_[words_train,categories]
documents_train[0]
```

```python
documents_test=words_test
```

```python
import random 
random.shuffle(documents_train)
random.shuffle(documents_test)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents_train=[(clean_review(document),category) for document,category in documents_train]
documents_train[0]
```

```python
documents_test=[(clean_review(document)) for document in documents_test]
documents_test[0]
```

```python
len(documents_train)
```

# Building Feature Set:

```python
training_document=documents_train[0:]
all_words=[]
for doc in training_document:
    all_words+=doc[0]
all_words
```

```python
testing_document=documents_test
```

```python
import nltk
freq=nltk.FreqDist(all_words)
freq
```

```python
common=freq.most_common(15)
common
```

```python
features=[i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in features:
        current_features[w]= w in words_set
    return current_features
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in training_document]
training_data
testing_data=[get_feature_dict(doc) for doc in testing_document]
testing_data
```

```python
training_data
```

# Naive Bayes:

```python
from nltk import NaiveBayesClassifier
```

```python
classifier=NaiveBayesClassifier.train(training_data)
```

```python
predictions=classifier.classify_many(testing_data)
predictions
```

```python
np.savetxt("predictions1.csv",predictions,fmt="%s") #0.5317
```

```python
testing_datas=np.c_[testing_data,predictions]
testing_datas
```

```python
nltk.classify.accuracy(classifier,testing_datas)
```

```python
classifier.show_most_informative_features(15)
```

# Count Vectorizer:

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
training_document[0]
```

```python
categories=[category for document ,category in training_document]
```

```python
text_documents=[" ".join(document) for document,category in training_document]
```

```python
text_documents[0]
```

```python
test_text_documents=[" ".join(document) for document in testing_document]
test_text_documents[0]
```

```python
count_vec=CountVectorizer(max_features=2000, ngram_range=(1,3),stop_words=stop)
```

```python
x_train_features=count_vec.fit_transform(text_documents)
```

```python
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features=count_vec.transform(test_text_documents)
x_test_features
```

```python
from sklearn.svm import SVC
```

```python
svc.fit(x_train_features,categories)
```

```python
predictions=svc.predict(x_test_features)
```

```python
np.savetxt("predictions4.csv",predictions,fmt="%s")
```

```python
svc.score(x_train_features,categories)
```
